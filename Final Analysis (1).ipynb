## Anova

```python
action_array = []
for action in ['Interval', 'Flowrate / VTBI Selection ', 'Channel Selection ', 'Drug Selection ']: 
    action_array.append(list(st.f_oneway(*[[data.durationData.loc[device, action]/1000000 for data in dataset] for device in ['Alaris', 'Baxter', 'ICU Medical', 'Ivenix']])))
pd.DataFrame(action_array, index=['Interval', 'Flowrate / VTBI Selection ', 'Channel Selection ', 'Drug Selection '], columns=['Coefficient', 'p_value'])
```

```python
action_array = []
devices = ['Alaris', 'Baxter', 'ICU Medical', 'Ivenix']
actions = ['Interval', 'Flowrate / VTBI Selection ', 'Channel Selection ', 'Drug Selection ']
for device in devices: 
    action_array.append(list(st.f_oneway(*[[data.durationData.loc[device, action]/1000000 for data in dataset] for action in actions])))
pd.DataFrame(action_array, index=devices, columns=['Coefficient', 'p_value'])
```

## AleexanderGovern

```python
action_array = []
devices = ['Alaris', 'Baxter', 'ICU Medical', 'Ivenix']
actions = ['Interval', 'Flowrate / VTBI Selection ', 'Channel Selection ', 'Drug Selection ']
for device in devices: 
    res = st.alexandergovern(*[[data.durationData.loc[device, action]/1000000 for data in dataset] for action in actions], nan_policy='omit')
    action_array.append([res.statistic, res.pvalue])
pd.DataFrame(action_array, index=devices, columns=['Coefficient', 'p_value'])
```

```python
def getAvg(db):
    rates = []
    for data in db:
        rates.append(data/np.array(data).sum())
    rate = rates[0]
    for r in rates:
        rate += r    
    return rate/31
```

```python
try:
    dbs = []
    for df in dataset:
        num = int(df.participant.split('Participant')[1].split('.')[0].split(' ')[0])
        if num in [9, 12, 13, 24, 25]: continue
        db = {'id':num}
        for name in ['Alaris', 'Baxter', 'ICU Medical', 'Ivenix']:
            for labl in ['Interval', 'Flowrate / VTBI Selection ', 'Channel Selection ', 'Drug Selection ']:
                if name=='Ivenix' and labl == 'Channel Selection ': continue
                frames =[]
                for i, j in zip(df.data.loc[ df.data['Event'].str.strip() == f'{name} {labl}Start' ].index, df.data.loc[ df.data['Event'].str.strip() == f'{name} {labl}End' ].index):
                    frames.append(df.data.iloc[i:j])
                db[name+" "+labl.strip()] = pd.concat(frames)
        dbs.append(db)
except:
    print(f'{num} {df.participant} -> {name} {len(df.data)}')
```

```python
bucket = []
for db in dbs:
    bucket.append(db['Baxter Interval'].loc[:,'Eye movement type'].value_counts())
# st.f_oneway(*bucket)
bucket
```

```python
table = []
devices = ['Alaris', 'Baxter', 'ICU Medical', 'Ivenix']
actions = ['Interval', 'Flowrate / VTBI Selection ', 'Channel Selection ', 'Drug Selection ']
for idx, device in enumerate(devices):
    table.append(list())
    for jdx, event in enumerate(actions):
        if device=='Ivenix' and event == 'Channel Selection ': table[idx].append(0);continue;
        d = pd.concat([db[f'{device} {event.strip()}'].loc[:,'Eye movement type'].value_counts() for db in dbs]).groupby(level=0).mean().round(1)*100
        table[idx].append((d/d.sum()).loc['Fixation'])
pd.DataFrame(table, columns=actions, index=devices).round(2).replace(0,'N/A')
```

```python
table = []
dtb = []
for device in ['Alaris', 'Baxter', 'ICU Medical', 'Ivenix']:
    tbl = []
    stb = []
    for event in ['Interval', 'Flowrate / VTBI Selection ', 'Channel Selection ', 'Drug Selection ']:
        if device=='Ivenix' and event == 'Channel Selection ': tbl = [0]; continue;
        ar = []
        sb = []
        for db in dbs:
            ddf = db[f'{device} {event.strip()}'].loc[:,'Eye movement type']
            sb.append(ddf)
            ar.append(ddf.value_counts())
        tbl.append(getAvg(ar).loc['Fixation'])
        stb.append([a.loc['Fixation'] for a in ar])
    table.append(tbl)
    dtb.append(stb)
pd.DataFrame(table, columns=['Interval', 'Flowrate / VTBI Selection', 'Channel Selection', 'Drug Selection'], index=['Alaris', 'Baxter', 'ICU Medical', 'Ivenix'])
```

```python
import scipy.stats as st
dta = [[ar for ar in stb] for stb in dtb]
coeff, b = st.f_oneway(*dta)
print(len(dta[0][0]))
```
